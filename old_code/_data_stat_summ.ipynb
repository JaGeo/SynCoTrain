```python
import numpy as np
import pandas as pd
import scipy as sc
import matplotlib.pyplot as plt
```

```python
posD = np.load("positive_data/local_crystal_features.npy", allow_pickle = True)
```

```python
theD = np.load("theoretical_data/local_crystal_features.npy", allow_pickle = True)
```

```python
p = posD[0]['feats']
p[0].keys()
```

```python
posD_cry_size = np.array([len(d['feats']) for d in posD])
theD_cry_size = np.array([len(d['feats']) for d in theD])
```

```python
plt.figure(figsize=(10,6))
plt.hist(posD_cry_size, bins=100)
plt.vlines(posD_cry_size.mean(),-20,1100, 'r',linestyles='--',  linewidth=3)
plt.vlines(np.median(posD_cry_size),-20,1100, 'y',linestyles='-',  linewidth=3)
plt.title('Experimental Crystal #Atoms');
```

```python
plt.figure(figsize=(10,6))
plt.hist(theD_cry_size, bins=100)
plt.vlines(theD_cry_size.mean(),-20,2100, 'r',linestyles='--',  linewidth=3)
plt.vlines(np.median(theD_cry_size),-20,2100, 'y',linestyles='-',  linewidth=3)
plt.title('Theoretical Crystal #Atoms');
```

```python
posD_element_num = np.array([len(set([(pp['element']) for pp in d['feats'].values()])) for d in posD])

theD_element_num = np.array([len(set([(pp['element']) for pp in d['feats'].values()])) for d in theD])
```

```python
plt.figure(figsize=(10,6))
plt.hist(posD_element_num, bins=10)
plt.vlines(theD_element_num.mean(),-100,5000, 'r',linestyles='--',  linewidth=5)
plt.vlines(np.median(theD_element_num),-100,5000, 'y',linestyles='-', linewidth=8)
plt.title('Experimental Crystal #Elements');
```

```python
plt.figure(figsize=(10,6))
plt.hist(theD_element_num, bins=10)
plt.vlines(theD_element_num.mean(),-100,15000, 'r',linestyles='--',  linewidth=5)
plt.vlines(np.median(theD_element_num),-100,15000, 'y',linestyles='-', linewidth=8)
plt.title('Theoretical Crystal #Elements');
```
