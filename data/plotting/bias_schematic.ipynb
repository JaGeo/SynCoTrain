```python
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic data
np.random.seed(0)  # for reproducibility
n_points = 100
x = np.linspace(0, 10, n_points)  # ground truth
noise = np.random.normal(0, 0.3, size=n_points)  # some noise
y = x + noise  # predicted values

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.6)
plt.plot([0, 10], [0, 10], 'r')  # line of perfect agreement
plt.xlabel('Ground Truth')
plt.ylabel('Predicted')
plt.title('Predicted vs Ground Truth')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic data
np.random.seed(0)  # for reproducibility
n_points = 1000
x = np.linspace(0, 10, n_points)  # ground truth
noise = np.random.normal(0, 0.25, size=n_points)  # some noise
# y = 0.9 * x + noise  # predicted values with a slight deviation in slope
y = 0.9 * x + noise +.5  # predicted values with a slight deviation in slope

# Add some outliers at low values
n_outliers = int(n_points * 0.05)
outlier_x = np.random.choice(x[100:n_outliers+100], size=n_outliers)  # choose random x values for outliers
outlier_y = outlier_x - 2 + np.random.normal(0, 0.2, size=n_outliers)  # corresponding y values for outliers

# Append outliers to original data
x = np.concatenate((x, outlier_x))
y = np.concatenate((y, outlier_y))

# Create scatter plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(x, y, alpha=0.6)
ax.plot([0, 10], [0, 10], 'r')  # line of perfect agreement
ax.set_xlabel('Ground Truth', fontsize = 25)
ax.set_ylabel('Predicted', fontsize = 25)
# ax.set_title('Predicted vs Ground Truth', fontsize = 25)
# ax.grid(True)

# Remove tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

# plt.show()
plt.savefig("bias_scheme.png", dpi = 350, transparent=True)
```
